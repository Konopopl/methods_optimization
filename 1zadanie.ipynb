## Задача 1
$1) $$f_1(\textbf{x}) = \|\textbf{A}\textbf{x} - \textbf{b}\|_2 + \|\textbf{x}\|_2$
Из прошлого семестра мы знаем, что норма является выпуклой функцией. Также, если $g(\textbf{x}) ~- $ выпуклая функция, то $\phi(\textbf{x}) = g(\textbf{A}\textbf{x} + \textbf{b})$ также является выпуклой $\forall \textbf{A}, \textbf{b}$. То есть $f_1(\textbf{x})$ выпукла. Из второй задачи мы знаем, что функция иметт два минимума, значит, она не является строго выпуклой.

$2)$ $f_2(\textbf{x}) = x_1^4 + x_2^4 + x_1^2 - 10x_1x_2 + 100x_2^2$. Гессиан данной функции равен $\nabla^2 f_2(\textbf{x}) = 12\begin{bmatrix} x_1^2 & 0 \\
0 & x_2^2 \end{bmatrix} + \begin{bmatrix} 2 & -10 \\
-10 & 200 \end{bmatrix}$ Найдём максимальное $m$, при котором матрица $\nabla^2 f_2(\textbf{x}) - m\textbf{E}$ всё ещё положительно полуопределена $\forall \textbf{x}$. Очевидно, это $m = 2$. То есть функция $f_2(\textbf{x})$ выпукла  и сильно выпукла.

$3)$ $f_3(\textbf{x}) = x_1^2 - 10x_1^2x_2^2 + 1000x_2^6$. $\nabla f_3 = [2x_1 - 20x_1x_2^2, -20x_1^2x_2 + 6000x_2^5]^T$. $\nabla^2 f_3 = \begin {bmatrix} 2 - 20x_2^2 && -40x_1x_2 \\ -40x_1x_2 && -20x_1^2 + 30000x_2^4 \end{bmatrix}$. Как видно, гессиан не является положительно полуопределённым, поскольку при $x_1 = 1, x_2 = 0$ имеет собственные значения разных знаков, то есть функция $f_3(\textbf{x})$ не является выпуклой.

$4)$ $f_4(\textbf{x}) = \frac{1}{1 + \exp{-x^2 + xy - y^2}} + x^2$. Возьмём вектора $[0, -1]^T$ и $[0, 1]^T$ и $\alpha = 0.5$. Тогда $f_4(0.5\cdot[0, -1]^T + 0.5\cdot[0, 1]^T) = f_4(\textbf{0}) = 1 > \frac{1}{1 + 1/e} = 0.5f_4([0, -1]^T) + 0.5f_4([0, 1]^T)$. Значит по определению функция $f_4$ не является выпуклой.

## Задача 2


$1)$ Уравнение $\textbf{0} = \nabla f(\textbf{x})$ не имееет решений, поэтому единственными критическими точками будут $(0, 0)$ и $\textbf{A}^{-1}\textbf{b} = (-333/40, 67/40)$. В силу выпуклости функции это могут быть только минимумы. Подставляя, $f(0, 0) = \sqrt{101}, f(-333/40, 67/40) = \frac{1}{40}\sqrt{115378} < \sqrt{101}$, то есть первая точка есть локальный минимум, а вторая - глобальный.

$2)$ $f_2(\textbf{x}) = x_1^4 + x_2^4 + x_1^2 - 10x_1x_2 + 100x_2^2$. Очевидно, градиент обнуляет точка $(0, 0)$. С другой стороны, поскольку функция сильно выпуклая, то других экструмумов нет, а $(0, 0)$ является глобальным минимумом.

$3)$ $f_3(\textbf{x}) = x_1^2 - 10x_1^2x_2^2 + 1000x_2^6$. Из равенства нулю градиента получаем пять точек $(0, 0)$, $(\pm\sqrt{3}, \mp\frac{1}{\sqrt{10}})$. В квадрате лежит только $(0, 0)$. При стремлении агрумента к бесконечности данная линейная дифференцируемая функция возрастает, поэтому у неё обязан быть глобальный минимум, причём среди этих пяти кандидатов. Прямой подстановкой в функцию убеждаемся, что это $(0, 0)$.

$4)$ Из равенства нулю градиента получаем единственного кандидата, что $x = 2y$, а также что $2x = 0$, то есть $(0,0)$ есть точка экстремума, а именно глобального минимума, поскольку гессиан $\begin{bmatrix} 2.5 & -0.25 \\
-0.25 & 0.5 \end{bmatrix}$ положительно определён.

```python
import numpy as np
import math

A = np.array([[1, -1], [20, 100]])
b = np.array([-10, 1])

def f1(x):
    x1, x2 = x
    ans = np.sqrt((x1 - x2 + 10)**2 + (20*x1 + 100*x2 - 1)**2) + np.sqrt(x1**2 + x2**2)
    return ans

def df1(x):
    x1, x2 = x
    f = np.dot(A.transpose(), np.dot(A, x) - b)/np.linalg.norm(np.dot(A, x) - b) + x/np.linalg.norm(x)
    return f

def f2(x):
    x1, x2 = x
    ans = np.power(x1, 4) + np.power(x2, 4) + np.power(x1, 2) - 10*x1*x2 + 100*np.power(x2, 2)
    return ans

def df2(x):
    x1, x2 = x
    f1 = 4*np.power(x1, 3) + 2*x1 - 10*x2
    f2 = 4*np.power(x2, 3) - 10*x1 + 200*x2
    f = np.array([f1, f2])
    return f

def f3(x):
    x1, x2 = x
    ans = x1**2 - (10*x1**2)*(x2**2) + 1000*x2**6
    return ans

def df3(x):
    x1, x2 = x
    f1 = 2*x1 - 20*x1*np.power(x2, 2)
    f2 = 20*np.power(x1, 2)*x2 + 6000*np.power(x2, 5)
    f = np.array([f1, f2])
    return f

def f4(x):
    x1, x2 = x
    ans = 1/(1 + math.e**(-x1**2-x2**2 + x1*x2)) + (x1)**2
    return ans

def df4(x):
    x1, x2 = x
    f1 = 2*x1 - 1/(1 + math.e**(-x1**2-x2**2 + x1*x2))**2*(math.e**(-x1**2-x2**2 + x1*x2))*(-2*x1 + x2)
    f2 = 1/(1 + math.e**(-x1**2-x2**2 + x1*x2))**2*(math.e**(-x1**2-x2**2 + x1*x2))*(-2*x2 + x1)
    f = np.array([f1, f2])
    return f
```

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
from matplotlib import pyplot as plt

def plot_3d_function(x1, x2, f, title, minima = None, iterations = None):
    '''
    '''
    low_lim_1 = x1.min()
    low_lim_2 = x2.min()
    up_lim_1 = x1.max()
    up_lim_2 = x2.max()
    X1, X2 = np.meshgrid(x1, x2)
    Z = f((X1, X2))

    fig = plt.figure(figsize=(16,7))
    fig.suptitle(title)

    ax = fig.add_subplot(1, 2, 1, projection='3d')
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    
    surf = ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,
    cmap=cm.RdBu,linewidth=0, antialiased=False)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
    if minima is not None:
        minima_ = np.array(minima).reshape(-1, 1)
        ax.plot(*minima_, f(minima_), 'r*', markersize=10)


    ax = fig.add_subplot(1, 2, 2)
    
    im = ax.imshow(Z,cmap=plt.cm.RdBu, origin = "lower", extent=[low_lim_1, up_lim_1, low_lim_2, up_lim_2])
    cset = ax.contour(X1, X2, Z,linewidths=2,cmap=plt.cm.Set2)
    ax.clabel(cset,inline=True,fmt='%1.1f',fontsize=10)
    fig.colorbar(im)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')

    if minima is not None:
        minima_ = np.array(minima).reshape(-1, 1)
        ax.plot(*minima_, 'r*', markersize=10)

    if iterations is not None:
        for point in iterations:
            ax.plot(*point, 'go', markersize=3)
        iterations = np.array(iterations).T
        ax.quiver(iterations[0,:-1], iterations[1,:-1], iterations[0,1:]-iterations[0,:-1], iterations[1,1:]-iterations[1,:-1], scale_units='xy', angles='xy', scale=1, color='blue')
    plt.savefig('plot_3d.svg')
    plt.show()
```

```python
def golden_search(f, a, b, epsilon):
    tau = (sqrt(5) + 1) / 2
    y = a + (b - a) / tau**2
    z = a + (b - a) / tau
    while b - a > epsilon:
        if f(y) <= f(z):
            b = z
            z = y
            y = a + (b - a) / tau**2
        else:
            a = y
            y = z
            z = a + (b - a) / tau
    return (a + b) / 2
```

```python
from scipy.optimize import minimize_scalar
from scipy.optimize import minimize
def steepest_descent(x_0, f, df, df_eps = 1e-8, max_iter = 1000):
    iterations = []
    norms = []
    x = np.array(x_0)
    iterations.append(x)
    norms.append(np.linalg.norm(df(x)))
    inner = 0
    oracle = 0
    while np.linalg.norm(df(x)) > df_eps and len(iterations) <= max_iter:
        a = 0
        b = 1
        epsilon = 1e-8
        tau = (math.sqrt(5) + 1) / 2
        y = a + (b - a) / tau**2
        z = a + (b - a) / tau
        oracle = oracle + 1
        while b - a > epsilon:
            oracle = oracle + 2
            inner = inner + 1
            if (f(x - y * df(x)) <= f(x - z * df(x))):
                b = z
                z = y
                y = a + (b - a) / tau**2
            else:
                a = y
                y = z
                z = a + (b - a) / tau
        res = (a + b) / 2
        # res = minimize_scalar(lambda alpha: f(x - alpha * df(x)), method ='Golden')        
        # res =  golden_search(lambda alpha: f(x - alpha * df(x)), 1/2, 1/2, 1e-4)
        # print('alpha', res)
        # alpha_opt = 0.01
        x = x - res * df(x)
        # print('norm ', np.linalg.norm(df(x)))
        iterations.append(x)
        norms.append(np.linalg.norm(df(x)))
    print('Finished with', len(iterations) - 1, 'iterations')
    print('Внутренних итераций ', inner)
    print('Вызовов оракула ', oracle)
    print('Norm ', np.linalg.norm(df(x)))
    print('x_N = ', iterations[len(iterations) - 1])
    print('Norm x_N = ', np.linalg.norm(iterations[len(iterations) - 1]))
    print('f_N = ', f(iterations[len(iterations) - 1]))
    ind = list(range(0, len(norms)))
    
    fig, ax = plt.subplots()

    ax.set_title('Точность по градиенту')
    
    plt.plot(ind, norms)
    return iterations
```

```python
def steepest_descent_auto(x_0, f, df, df_eps = 1e-18, max_iter = 1000):
    iterations = []
    norms = []
    x = np.array(x_0)
    iterations.append(x)
    norms.append(np.linalg.norm(df(x)))
    while np.linalg.norm(df(x)) > df_eps and len(iterations) <= max_iter:
        res = minimize_scalar(lambda alpha: f(x - alpha * df(x)), method ='Golden', bounds = (0, 1))        
        # res =  golden_search(lambda alpha: f(x - alpha * df(x)), 1/2, 1/2, 1e-4)
        # print('alpha ', res.x)
        # alpha_opt = 0.01
        # alp
        x = x - res.x * df(x)
        # print('norm ', np.linalg.norm(df(x)))
        iterations.append(x)
        norms.append(np.linalg.norm(df(x)))
    print('Finished with', len(iterations) - 1, 'iterations')
    print('Norm ', np.linalg.norm(df(x)))
    print('x_N = ', iterations[len(iterations) - 1])
    print('Norm x_N = ', np.linalg.norm(iterations[len(iterations) - 1]))
    print('f_N = ', f(iterations[len(iterations) - 1]))
    ind = list(range(0, len(norms)))
    
    fig, ax = plt.subplots()

    ax.set_title('Точность по градиенту')
    
    plt.plot(ind, norms)
    
    return iterations
```

```python
def reverse_descent(x_0, f, df, df_eps = 1e-8, max_iter = 1000):
    iterations = []
    norms = []
    x = np.array(x_0)
    iterations.append(x)
    norms.append(np.linalg.norm(df(x)))
    oracle = 0
    inner = 0
    while np.linalg.norm(df(x)) > df_eps and len(iterations) <= max_iter:
        it = 0
        u1 = 1e-9
        u2 = 0.9
        p1 = 0.5
        p2 = 1.5
        alpha = 1
        while(inner < 330000):
            inner = inner + 1
            oracle = oracle + 2
            if (alpha*u1*np.dot(df(x).transpose(), df(x)) > f(x) - f(x - alpha * df(x))):
                alpha = p1*alpha
                it = it + 1
            else:
                if (f(x) - f(x - alpha * df(x)) > alpha*u2*np.dot(df(x).transpose(), df(x))):
                    alpha = alpha*p2
                    it = it + 1
                else:
                    break

        # res = minimize_scalar(lambda alpha: f(x - alpha * df(x)), method ='Brent')        
        # res =  golden_search(lambda alpha: f(x - alpha * df(x)), 1/2, 1/2, 1e-4)
        # print('alpha ',alpha)
        # alpha_opt = 0.01
        # alp
        x = x - alpha * df(x)
        # print('norm ', np.linalg.norm(df(x)))
        iterations.append(x)
        norms.append(np.linalg.norm(df(x)))
    print('Finished with', len(iterations) - 1, 'iterations')
    print('Внутренних итераций ', inner)
    print('Вызовов оракула ', oracle)
    print('Norm ', np.linalg.norm(df(x)))
    print('x_N = ', iterations[len(iterations) - 1])
    print('Norm x_N = ', np.linalg.norm(iterations[len(iterations) - 1]))
    print('f_N = ', f(iterations[len(iterations) - 1]))
    ind = list(range(0, len(norms)))
    
    fig, ax = plt.subplots()

    ax.set_title('Точность по градиенту')
    
    plt.plot(ind, norms)
    return iterations
```
